# 贞人占卜 — 一键编译 APK

本 Notebook 在 Google Colab 中运行，自动将 Python 源码编译为 Android APK。

### 使用步骤
1. 点击「文件」→「上传」，上传以下 3 个文件到 Colab：
   - `main.py`
   - `zhenren.kv`
   - `oracle_engine.py`
   - `buildozer.spec`
2. 依次运行下方代码块
3. 编译完成后自动下载 APK（约需 15-25 分钟）

```python
# 第一步：安装系统依赖
!sudo apt update -qq
!sudo apt install -y -qq python3-pip git zip unzip openjdk-17-jdk autoconf libtool pkg-config zlib1g-dev libncurses5-dev libncursesw5-dev libtinfo5 cmake libffi-dev libssl-dev automake libltdl-dev
```

```python
# 第二步：安装 Buildozer
!pip install buildozer cython
```

```python
# 第三步：确认文件完整
!ls -la main.py zhenren.kv oracle_engine.py buildozer.spec
!mkdir -p .buildozer
```

```python
# 第四步：初始化 Buildozer（首次运行）
# 如果已上传 buildozer.spec，运行此步确认配置
!buildozer init 2>/dev/null
!echo '确认 spec 文件:'; cat buildozer.spec | head -5
```

```python
# 第五步：编译 APK（耗时 15-25 分钟）
!buildozer -v android debug
```

```python
# 第六步：下载 APK
from google.colab import files
import glob

apk_files = glob.glob('bin/*.apk')
if apk_files:
    print(f'编译成功！APK 文件: {apk_files[0]}')
    files.download(apk_files[0])
else:
    print('未找到 APK 文件，请检查上一步的编译日志。')
    !find . -name '*.apk' 2>/dev/null
```
